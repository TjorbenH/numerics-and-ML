# Double Pendulum 

## The Problem
#### Assumptions
We are looking at a double pendulum with two limbs that are massless and of length $l_1$ and $l_2$. Attached at the bottom of those limbs are two masses $m_1$ and $m_2$. We will only consider pendulums that move in the 2D plane. 

#### Position of the Pendulums 
The position of the two masses will be described according to the angle of the limbs to the vertical. We can denote the positions of the masses according to those two angles $\theta_1$ and $\theta_2$. 
$$
\begin{pmatrix}
x_1 \\ y_1
\end{pmatrix}
=
\begin{pmatrix}
l_1\sin(\theta_1) \\ -l_1\cos(\theta_1)
\end{pmatrix}
$$

$$
\begin{pmatrix}
x_2 \\ y_2
\end{pmatrix}
=
\begin{pmatrix}
l_1\sin(\theta_1) + l_2\sin(\theta_2) \\ -l_1\cos(\theta_1) - l_2\cos(\theta_2)
\end{pmatrix}
$$

### Equations of Motion
In the end the derivations yield a nonlinear system of ordinary second order differential equations. It is impossible to solve this system analytically, which is why we are here in the first place. 

$$
\ddot{\theta_1} = - \frac{m_2}{m_1+m_2}\frac{l_2}{l_1}\left(\ddot{\theta_2}\cos(\theta_1-\theta_2)+\dot{\theta_2}^2\sin(\theta_1-\theta_2)\right)-\frac{g}{l_1}\sin(\theta_1)
$$

$$
\ddot{\theta_2} = - \frac{l_1}{l_2}\left(\ddot{\theta_1}\cos(\theta_1-\theta_2)-\dot{\theta_1}^2\sin(\theta_1-\theta_2)\right)-\frac{g}{l_2}\sin(\theta_2)
$$
